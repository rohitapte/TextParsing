```python
import nltk
nltk.download()
from nltk import pos_tag, word_tokenize,sent_tokenize
from collections import Counter
```

```python
from nltk.corpus import PlaintextCorpusReader
#corpus_root=r"C:\Users\tihor\OneDrive\Documents\Python Scripts\TextParsing\FOMC"
corpus_root=r"C:\Users\user\OneDrive\Documents\Python Scripts\TextParsing\FOMC"
fed=PlaintextCorpusReader(corpus_root, '.*')
fed.fileids()
```

```python
modals=['','','unemployment','inflation']
searchwords={}
for file in fed.fileids():
    wordcounts=Counter([w.lower() for w in fed.words(file)])
    words=[]
    words.append(wordcounts["activity"])
    words.append(wordcounts["labor"])
    words.append(wordcounts["unemployment"])
    words.append(wordcounts["inflation"])
    words.append(wordcounts["conditions"])
    searchwords[file]=words
    print(fed.sents(file).collocations())
#print(searchwords)
```
